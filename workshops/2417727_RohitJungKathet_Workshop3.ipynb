```python
total_fare = titanic_df['Fare'].sum()
total_fare
```

```python
second_class = titanic_df.loc[titanic_df['Pclass'] == 2]
second_class_fare = second_class['Fare'].sum()
second_class_fare
```

```python
third_class = titanic_df.loc[titanic_df['Pclass'] == 3]
third_class_fare = third_class['Fare'].sum()
third_class_fare
```

```python
#proportion of fare revenue by class
proportions = np.array(fare_list) / total_fare
proportions
```

```python
titanic_df.dropna(subset=['Age'], inplace= True)

# titanic_df['Age']= titanic_df['Age'].dropna()
titanic_df['Age'].isnull().sum()
titanic_df['Age'].sample(10)
```

```python
titanic_df['age_group'] = pd.cut(titanic_df['Age'], bins=[0, 17, 64, float('inf')], labels=['Child', 'Adult', 'Senior'])
```

```python
total_passengers = titanic_df['PassengerId'].count()
total_passengers
```

```python
age_group_values = titanic_df['age_group'].value_counts()
age_group_values
```

```python
prop = age_group_values / total_passengers
prop
```

```python
prop_percentage = prop * 100
prop_percentage
```
